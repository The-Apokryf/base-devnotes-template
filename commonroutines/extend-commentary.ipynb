```python
formator = lambda digEnt : [
    digEnt.model, 
    f'{(math.floor(digEnt.size/1024/1024)/1000)}GB', 
    digEnt.modified_at.strftime("%Y-%m-%d"),
    digEnt.details.format,
    digEnt.details.family,
    digEnt.details.parent_model,
    digEnt.details.parameter_size,
    digEnt.details.quantization_level
]

listing = pandas.DataFrame([formator(digEnt) for digEnt in population.models])

with open("list-models.html", "w") as listprint:
    listprint.write(listing.to_html())
```
